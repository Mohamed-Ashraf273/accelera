```python
import json
import pickle

from accelera.src.code_parallelizer.utils.code_utils import extract_features
from accelera.src.code_parallelizer.utils.code_utils import (
    feature_dict_to_vector,
)
from accelera.src.code_parallelizer.utils.code_utils import generate_omp_pragma
```

```python
le = pickle.load(
    open(
        "/home/mohamed-ashraf/Desktop/projects/accelera/accelera/src/code_parallelizer/models/label_encoder.pkl",
        "rb",
    )
)
model = pickle.load(
    open(
        "/home/mohamed-ashraf/Desktop/projects/accelera/accelera/src/code_parallelizer/models/model.pkl",
        "rb",
    )
)
```

```python
data = []
with open(
    "/home/mohamed-ashraf/Desktop/projects/accelera/accelera/src/code_parallelizer/data/data.json",
    "r",
) as f:
    for line in f:
        line = line.strip()
        obj = json.loads(line)
        data.append(obj)

print(data[0]["code"])
print(data[0]["pragma"])
```

```python
code = data[900]["code"]
features = extract_features(code)
embedding = feature_dict_to_vector(features)

print(f"Embedding shape before reshape: {embedding.shape}")
print(f"Embedding type: {type(embedding)}")

embedding_2d = embedding.reshape(1, -1)
print(f"Embedding shape after reshape: {embedding_2d.shape}")

pred = model.predict(embedding_2d)
pred_class = le.inverse_transform(pred)
predicted_pragma = generate_omp_pragma(code, pred_class[0])

print(f"\nPredicted class index: {pred[0]}")
print(f"Predicted class name: {pred_class[0]}")
print(f"Actual pragma: {data[0]['pragma']}")
print(f"Generated pragma: \n{predicted_pragma}")
```
